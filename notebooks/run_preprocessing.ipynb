```python
data = pd.read_csv(join(PATH_DATA, "arabic_data.csv"), names=['text', 'label'], engine="python", skiprows=1)
pattern = "(?:[A-Za-z])"
data['arabic'] = data['text'].str.contains(pattern)
data['text'] = data['text'].astype(str)
texts = []
[texts.append(preprocess(text)) for text in data['text'].values.tolist() if text]
data['text'] = texts
data = data.drop_duplicates(subset ="text", keep = 'first')
data = data.dropna(subset=['text'])
data = data.dropna(subset=['text'])
data = data.sample(frac=1).reset_index(drop=True)
arabic = data.loc[data['arabic'] == False]
arabic
```
