```python
%load_ext autoreload
%autoreload 2


import pandas as pd

import spacy

import random


from os import getcwd
from os.path import join, dirname

PATH_REPO = dirname(getcwd())
PATH_UTILS = join(PATH_REPO, 'utils')
PATH_DATA = join(PATH_REPO, 'data')
PATH_MODELS = join(PATH_REPO, 'models')

import _pickle as cPickle

import sys
sys.path.append(PATH_UTILS)

pd.set_option('max_colwidth', None)

random.seed(1)
spacy.util.fix_random_seed(1)


from training import  make_textcat_predictions

from sklearn.metrics import precision_recall_fscore_support as score

nlp = spacy.load(join(PATH_MODELS, 'gratitude_categorizer_3'))
```

```python
test = pd.read_csv(join(PATH_DATA, 'dev_set.csv'), engine="python")
# test['label'] = np.where(test['label']=='Gratitude', 1, 0)
```

```python
scores = make_textcat_predictions(test, nlp)
test['score'] = scores
test
```

```python
predictions = []
for s in scores:
    if s > 0.6:
        predictions.append(1)
    else:
        predictions.append(0)
test['prediction'] = predictions
```

```python
true = test['label'].values.tolist()
test
```

```python
precision, recall, fscore, support = score(true, predictions, average='macro')
print("Precision:\t", precision)
print("Recall:\t", recall)
print("F1:\t", fscore)
```
